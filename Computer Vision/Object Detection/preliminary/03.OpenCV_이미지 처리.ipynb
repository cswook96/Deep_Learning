```python
# OpenCV로 이미지 로드
import cv2
import matplotlib.pyplot as plt


cv2_image = cv2.imread('/content/data/beatles01.jpg')
cv2.imwrite('/content/data/beatles02_cv.jpg', cv2_image)

fig, ax = plt.subplots(figsize=(10, 10))
img = plt.imread('/content/data/beatles02_cv.jpg')
ax.imshow(img)

plt.show()
```

```python
# OpenCV의 imread()로 반환된 BGR 이미지 넘파이 배열을 그대로 시각화 하기
cv2_image = cv2.imread('/content/data/beatles01.jpg')

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(cv2_image)

plt.show()
```

```python
# BGR -> RGB
cv2_image = cv2.imread('/content/data/beatles01.jpg')
draw_image = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(2, 1, figsize=(10, 20))
axes[0].imshow(draw_image)

# RGB를 cv2,imwrite를 이용하여 저장한 이미지
cv2.imwrite('/content/data/beatles03_cv.jpg', draw_image)
img = plt.imread('/content/data/beatles03_cv.jpg')
axes[1].imshow(img)

plt.show()
```
