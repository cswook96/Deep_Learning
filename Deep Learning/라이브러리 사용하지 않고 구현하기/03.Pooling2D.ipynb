## AveragePooling2D

```python
import random

pool_size = 3
H, W = 7, 8
data = []

for _ in range(H):
    data_row = [random.randint(0,5) for _ in range(W)]
    data.append(data_row)
print('Input data:')
for row in data:
    print(row)
print('\n')

valid_H = H - pool_size + 1
valid_W = W - pool_size + 1

pooled = []
for h in range(valid_H):
    pooled_vec = []
    for w in range(valid_W):
        window = []
        for inner_h in range(pool_size):
            window_vec = data[h+inner_h][w:w+pool_size]
            window.append(window_vec)
        
        window_sum = 0
        for inner_h in range(pool_size):
            for inner_w in range(pool_size):
                window_sum += window[inner_h][inner_w]
        window_mean = window_sum / pool_size**2
        pooled_vec.append(window_mean)
    pooled.append(pooled_vec)

print('Pooling data:')
for row in pooled:
    print(row)
```

## MaxPooling2D

```python
import random

pool_size = 3
H, W = 6, 6
data = []
for _ in range(H):
    data_vec = [random.randint(0,100) for _ in range(W)]
    data.append(data_vec)
print('Input data:')
for row in data:
    print(row)
print('\n')

valid_H = H - pool_size + 1
valid_W = W - pool_size + 1

pooled = []
for h in range(valid_H):
    pooled_vec = []
    for w in range(valid_W):
        window = []
        for inner_h in range(pool_size):
            window_vec = data[h+inner_h][w:w+pool_size]
            window.append(window_vec)

        window_M = None
        for inner_h in range(pool_size):
            for inner_w in range(pool_size):
                if window_M == None or window[inner_h][inner_w] > window_M:
                    window_M = window[inner_h][inner_w]
        pooled_vec.append(window_M)
    pooled.append(pooled_vec)

print('pooling data:')
for row in pooled:
    print(row)
```
