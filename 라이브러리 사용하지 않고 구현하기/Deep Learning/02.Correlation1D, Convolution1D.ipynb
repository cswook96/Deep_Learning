## Correlation
<br>
<font size="5">
$$ \vec{x} \otimes \vec{f} [i] = \sum_{u=-k}^{k} x[i+u] f[u]$$
</font>


<br>
<font size="5">
    $$ L_{o} = L_{i} - L_{f} + 1 $$
</font>

```python
import random
import matplotlib.pyplot as plt

data = [random.randint(-1,1) for _ in range(30)]
corr_filter = [-1,1,1]
data_len = len(data)
filter_len = len(corr_filter)
print('Input data:',data)
print('Length of data:',data_len)

valid_len = data_len - filter_len + 1

correlation = []
for filter_loc in range(valid_len):
    window = data[filter_loc : filter_loc + filter_len]
    
    corr = 0
    for w_val, f_val in zip(window, corr_filter):
        corr += w_val*f_val
    correlation.append(corr)

print('Correlation:',correlation)
print('Length of Correlation:',len(correlation))
print('\n')

fig, ax = plt.subplots(figsize=(10,5))
correlation = [0] + correlation
ax.plot(data,label= 'data')
ax.plot(correlation,label='correlation value')
ax.legend(fontsize=10)
ax.grid()
```

## Convolution
<br>
<font size="5">
$$ \vec{x} * \vec{f} [i] = \sum_{u=-k}^{k} x[i+u] f[-u]$$
</font>


<br>
<font size="5">
    $$ L_{o} = L_{i} - L_{f} + 1 $$
</font>

```python
import random
import matplotlib.pyplot as plt

data = [random.randint(-1,1) for _ in range(30)]
data_len = len(data)
conv_filter = [-1,1,1]
filter_len = len(conv_filter)
print('Input data:',data)
print('Length of Input data:',data_len)

valid_len = data_len - filter_len + 1

convolution = []
for filter_loc in range(valid_len):
    window = data[filter_loc : filter_loc + filter_len]
    
    conv = 0
    for w_val,f_val in zip(window,conv_filter[::-1]):
        conv += w_val*f_val
    convolution.append(conv)

print('Convolution:',convolution)
print('Length of Convolution:',len(convolution))
print('\n')

fig, ax = plt.subplots(figsize=(10,5))
convolution = [0] + convolution
ax.plot(data, label = 'data')
ax.plot(convolution, label = 'Convolution value')
ax.legend(fontsize=10)
ax.grid()
```
