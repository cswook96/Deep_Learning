## Correlation2D


```python
import random

filter_size = 3
H, W = 5, 6

data = []
for _ in range(H):
    data.append([random.randint(0,10) for _ in range(W)])
print('Input data:')
for data_row in data:
    print(data_row)
print('\n')

corr_filter = []
for _ in range(filter_size):
    corr_filter.append([random.randint(0,10) for _ in range(filter_size)])
print('Correlational Filter:')
for filter_row in corr_filter:
    print(filter_row)
print('\n')

valid_H = H - filter_size + 1
valid_W = W - filter_size + 1

correlation = []
for h in range(valid_H):
    correlation_vec = []
    for w in range(valid_W):
        window = []
        for inner_h in range(filter_size):
            window_vec = data[h+inner_h][w:w+filter_size]
            window.append(window_vec)

        corr = 0
        for window_vec, filter_vec in zip(window,corr_filter):
            for window_val, filter_val in zip(window_vec, filter_vec):
                corr += window_val * filter_val
        correlation_vec.append(corr)
    correlation.append(correlation_vec)

print('Correlation:')
for corr_row in correlation:
    print(corr_row)
```

## Convolution2D

```python
# method 1

import random

filter_size = 3
H, W = 5, 6

data = []
for _ in range(H):
    data.append([random.randint(0,10) for _ in range(W)])
print('Input data:')
for data_row in data:
    print(data_row)
print('\n')

conv_filter = []
for _ in range(filter_size):
    conv_filter.append([random.randint(0,10) for _ in range(filter_size)])
print('Convolutional Filter:')
for filter_row in conv_filter:
    print(filter_row)
print('\n')

valid_H = H - filter_size + 1
valid_W = W - filter_size + 1

convolution = []
for h in range(valid_H):
    convolution_vec = []
    for w in range(valid_W):
        window = []
        for inner_h in range(filter_size):
            window_vec = data[h+inner_h][w:w+filter_size]
            window.append(window_vec)

        conv = 0
        for window_vec, filter_vec in zip(window,conv_filter[::-1]):
            for window_val, filter_val in zip(window_vec, filter_vec[::-1]):
                conv += window_val * filter_val
        convolution_vec.append(conv)
    convolution.append(convolution_vec)

print('Convolution:')
for conv_row in convolution:
    print(conv_row)
```
