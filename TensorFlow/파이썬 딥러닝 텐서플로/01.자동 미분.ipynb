```python
import tensorflow as tf
import numpy as np
```

```python
g = tf.random.Generator.from_seed(2020)
X = g.normal(shape=(10,))
Y = 3*X -2
```

```python
# mse 계산
def cal_mse(X,Y,a,b):
    Y_pred = a*X + b
    se = (Y - Y_pred)**2
    mse = tf.reduce_mean(se)

    return mse
```

```python
# 자동 미분
a = tf.Variable(0.0)
b = tf.Variable(0.0)

EPOCHS = 200

for epoch in range(1,EPOCHS + 1):

    with tf.GradientTape() as tape:
        mse = cal_mse(X,Y,a,b)
        
        grad = tape.gradient(mse, {'a':a, 'b':b})
        d_a, d_b = grad['a'], grad['b']

        a.assign_sub(d_a*0.05) # 0.05:learning rate
        b.assign_sub(d_b*0.05)
        
        if epoch % 20 == 0:
            print('EPOCH %d - MSE:%4f --a:%2f --b:%2f'%(epoch, mse, a, b))
```
