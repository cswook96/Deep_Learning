```python
import tensorflow as tf
import numpy as np
```

```python
# load datasets
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_valid, y_valid) = mnist.load_data()
```

```python
# normalization
x_train = (x_train/255.0).astype(np.float32)
x_valid = (x_valid/255.0).astype(np.float32)
```

```python
# modeling
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(256, activation='relu'),
                                    tf.keras.layers.Dense(64, activation='relu'),
                                    tf.keras.layers.Dense(32, activation='relu'),

                                    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# compile
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
```

```python
# ReduceLROnPlateau 콜백 생성
reduceLR = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', # val_loss 기준 
                                                factor=0.5,         # callback 호출시 학습률을 1/2로 줄임
                                                patience=3,         # epoch 10동안 개선되지 않으면 callback 호출
                                                verbose=1)
```

```python
model.fit(x_train, y_train,
          validation_data=(x_valid, y_valid),
          shuffle=True,
          batch_size=128, validation_batch_size=32,
          epochs=20,
          callbacks=[reduceLR])
```
