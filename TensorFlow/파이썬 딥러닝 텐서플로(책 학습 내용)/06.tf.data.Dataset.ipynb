```python
import numpy as np
import tensorflow as tf
```

```python
# as_numpy_iterator(넘파이 배열로 변환)
dataset = tf.data.Dataset.range(10)
dataset_np = dataset.as_numpy_iterator()
print(f"{dataset}, type: {type(dataset)}")
print(f"{dataset_np}, type: {type(dataset_np)}")
```

```python
# apply(5보다 작은수만 리턴)
dataset = tf.data.Dataset.range(10)

def filter_five(x):
    return x.filter(lambda x: x<5)

list(dataset.apply(filter_five).as_numpy_iterator())
```

```python
# from_tensor_slices(Dataset 클래스로 변환)
ds = tf.data.Dataset.from_tensor_slices([1, 2, 3, 4, 5])
print(type(ds))
list(ds.as_numpy_iterator())
```

```python
# iterattion
ds = tf.data.Dataset.from_tensor_slices([1,2,3,4,5])

for d in ds:
    print(d)
print()

for d in ds.take(3):
    print(d)
```

```python
# range
df = tf.data.Dataset.range(1, 10, 2)

for d in ds:
    print(d)
```

```python
# batch
ds = tf.data.Dataset.range(8)

for d in ds.batch(3):
    print(d)
```

```python
# dror remainder
ds = tf.data.Dataset.range(8)
for d in ds.batch(3, drop_remainder=True):
    print(d)
```

```python
# window
ds = tf.data.Dataset.range(10)
ds = ds.window(5, shift=2, drop_remainder=False) # drop_remainder=False
for d in ds:
    print(list(d.as_numpy_iterator()))
print()

ds = tf.data.Dataset.range(10)
ds = ds.window(5,shift=2, drop_remainder=True) # drop_remainder=True(일반적)
for d in ds:
    print(list(d.as_numpy_iterator()))
```

```python
# flat_map
ds = tf.data.Dataset.range(10)
ds = ds.window(5, shift=1, drop_remainder=True) # drop_remainder=False
for d in ds:
    print(d)
print()

ds = tf.data.Dataset.range(10)
ds = ds.window(5, shift=1, drop_remainder=True)
ds = ds.flat_map(lambda w:w.batch(5))
for d in ds:
    print(d)
```

```python
# shuffle
ds = tf.data.Dataset.from_tensor_slices(np.arange(10))

for d in ds:
    print(d)
print()

ds = tf.data.Dataset.from_tensor_slices(np.arange(10)).shuffle(buffer_size=5)
for d in ds:
    print(d)
```

```python
# map
window_size=5

ds = tf.data.Dataset.range(10)
ds = ds.window(window_size, shift=1, drop_remainder=True)
ds = ds.flat_map(lambda w: w.batch(window_size))
ds = ds.shuffle(10)

# 첫 4개와 마지막 1개를 분리
ds = ds.map(lambda w:(w[:-1],w[-1]))
for x, y in ds:
    print(f"train set: {x}\nlabel set:{y}")
    print('==='*10)
```
