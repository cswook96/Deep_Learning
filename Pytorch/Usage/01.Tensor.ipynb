## 1.Making Tensor

```python
import torch
import numpy as np

x = torch.empty(5, 4) # 5x4 임의의 텐서 
print(f"tensor:\n{x}, shape:{x.shape}\n")

ones = torch.ones((3,3)) # 3x3 1 텐서
print(f"ones:\n{ones}, shape:{ones.shape}\n")

zeros = torch.zeros(4) # 4개 0 텐서
print(f"zeros:\n{zeros}, shape:{zeros.shape}\n")

rand = torch.rand(5, 6) # 5x6 랜뎀 텐서(표준 정규분포를 기준으로 생성)
print(f"rand:\n{rand}, shape:{rand.shape}")
```

## 2.List,ndarray -> Tensor

```python
import torch
import numpy as np

l = [1,3,5] # list
ar = np.array([11,12,13]) # ndarray
print(f"list: {l}")
print(f"ndarray: {ar}\n")

te1 = torch.tensor(l) # list -> tensor
te2 = torch.tensor(ar) # ndarray -> tensor
print(f"list -> tensor: {te1}")
print(f"ndarray -> tensor: {te2}\n")

# tensor -> list
ar1 = np.array(te1) 
ar2 = te2.numpy()
print(f"ar1:\n{ar1}")
print(f"ar2:\n{ar2}")
```

## 3.size,type of Tensor

```python
import torch
import numpy as np

x = np.random.randint(0, 10, (3, 4))
tensor = torch.tensor(x)

# size
print(tensor.size())

# type
print(f"type: {type(tensor)}")
```

## 4.addition of Tensor

```python
import torch

x = torch.rand(2, 2)
y = torch.rand(2, 2)
print(f"x:\n{x}")
print(f"y:\n{y}\n")

print(f"x + y:\n{x + y}")
print(f"torch.add(x, y):\n{torch.add(x, y)}")
print(f"x.add(y):\n{x.add(y)}\n")

print(f"x.add_(y):\n{x.add_(y)}") # inplace
print(f"x:\n{x}")
```

## 5.Changing size of Tensor

```python
import torch

x = torch.rand((8, 8))
print(f"x.size: {x.size()}\n")

# view
print(f"x.view(64): {x.view(64).size()}")
print(f"x.view(-1, 4, 4): {x.view(-1, 4, 4).size()}")
```

## 6.단일 텐서에서 값으로 뽑아내기

```python
import torch

x = torch.ones(1, dtype=int)

print(f"x: {x}")
print(f"x.item(): {x.item()}")
```
